```python
import pandas as pd


df = pd.read_excel("E:\\mp\\img\\new adidas.xlsx")


# Convert retailer names to lowercase for case-insensitive comparison
df["Retailer"] = df["Retailer"].str.lower()

# Check unique retailer names
unique_retailers = df["Retailer"].unique()
print(unique_retailers)
# Remove leading and trailing spaces from retailer names
df["Retailer"] = df["Retailer"].str.strip()

# Write the modified DataFrame back to a CSV file
df.to_excel("new adidas.xlsx", index=False)


```
