```python
import pandas as pd
import numpy as np
import random

columns = ['InvoiceDate', 'Retailer', 'Region', 'State', 'City', 'Product', 'Price per Unit',
           'UnitsSold', 'TotalSales', 'OperatingProfit', 'OperatingMargin', 'SalesMethod']

regions = ['Northeast', 'South', 'West', 'Midwest', 'Southeast']
states_by_region = {
    'Northeast': ['New York', 'Pennsylvania', 'Maine', 'West Virginia', 'Maryland', 'Delaware', 'New Jersey', 'Connecticut', 'Rhode Island', 'Massachusetts', 'Vermont', 'New Hampshire'],
    'South': ['Texas', 'Tennessee', 'Alabama', 'Louisiana', 'Mississippi', 'Arkansas', 'Oklahoma'],
    'West': ['California', 'Nevada', 'Colorado', 'Washington', 'Alaska', 'Hawaii', 'Wyoming', 'Utah', 'Oregon', 'Idaho', 'Arizona', 'New Mexico'],
    'Midwest': ['Illinois', 'Minnesota', 'Montana', 'Nebraska', 'Michigan', 'Missouri', 'Ohio', 'Kansas', 'South Dakota', 'North Dakota', 'Iowa', 'Wisconsin', 'Indiana'],
    'Southeast': ['Florida', 'Virginia', 'Georgia', 'South Carolina', 'North Carolina', 'Kentucky']
}
cities_by_state = {
    'Northeast': {
        'New York': ['New York', 'Albany'],
        'Pennsylvania': ['Philadelphia'],
        'Maine': ['Portland'],
        'West Virginia': ['Charleston'],
        'Maryland': ['Baltimore'],
        'Delaware': ['Wilmington'],
        'New Jersey': ['Newark'],
        'Connecticut': ['Hartford'],
        'Rhode Island': ['Providence'],
        'Massachusetts': ['Boston'],
        'Vermont': ['Burlington'],
        'New Hampshire': ['Manchester']
    },
    'South': {
        'Texas': ['Houston', 'Dallas'],
        'Tennessee': ['Knoxville'],
        'Alabama': ['Birmingham'],
        'Louisiana': ['New Orleans'],
        'Mississippi': ['Jackson'],
        'Arkansas': ['Little Rock'],
        'Oklahoma': ['Oklahoma City']
    },
    'West': {
        'California': ['San Francisco', 'Los Angeles'],
        'Nevada': ['Las Vegas'],
        'Colorado': ['Denver'],
        'Washington': ['Seattle'],
        'Alaska': ['Anchorage'],
        'Hawaii': ['Honolulu'],
        'Wyoming': ['Cheyenne'],
        'Utah': ['Salt Lake City'],
        'Oregon': ['Portland'],
        'Idaho': ['Boise'],
        'Arizona': ['Phoenix'],
        'New Mexico': ['Albuquerque']
    },
    'Midwest': {
        'Illinois': ['Chicago'],
        'Minnesota': ['Minneapolis'],
        'Montana': ['Billings'],
        'Nebraska': ['Omaha'],
        'Michigan': ['Detroit'],
        'Missouri': ['St. Louis'],
        'Ohio': ['Columbus'],
        'Kansas': ['Wichita'],
        'South Dakota': ['Sioux Falls'],
        'North Dakota': ['Fargo'],
        'Iowa': ['Des Moines'],
        'Wisconsin': ['Milwaukee'],
        'Indiana': ['Indianapolis']
    },
    'Southeast': {
        'Florida': ['Miami', 'Orlando'],
        'Virginia': ['Richmond'],
        'Georgia': ['Atlanta'],
        'South Carolina': ['Charleston'],
        'North Carolina': ['Charlotte'],
        'Kentucky': ['Louisville']
    }
}

Product = ["Men's Apparel",
           "Men's Athletic Footwear",
           "Men's Street Footwear",
           "Women's Apparel",
           "Women's Athletic Footwear",
           "Women's Street Footwear"]

SalesMethod = ['Online', 'Outlet', 'In-store']

Retailers = ['Amazon', 'Foot Locker', "Kohl's", 'Sports Direct', 'Walmart', 'West Gear']

dummy_data_rows = []

for date in pd.date_range(start='2019-01-01', end='2019-12-31', freq='D'):  # Adjusted date range
    for _ in range(25):  # Generate 25 records per day
        region = random.choice(regions)
        state = random.choice(states_by_region[region])
        city = random.choice(cities_by_state[region][state])  # Adjusted to select city based on region and state
        product = random.choice(Product)
        retailer = random.choice(Retailers)
        if retailer == 'Amazon':
            sales_method = 'Online'
        else:
            sales_method = random.choice(SalesMethod)
        price_per_unit = round(np.random.uniform(30, 70) / 5) * 5  # Adjusted to round price to nearest multiple of 5
        units_sold = np.random.randint(1000, 2001)  # Adjusted units sold range
        total_sales = round(price_per_unit * units_sold, 2)
        operating_margin = round(np.random.uniform(0.3, 0.6), 2) * 100  # Adjusted to generate margin percentage between 30-60%
        operating_profit = round(total_sales * operating_margin / 100, -1)  # Adjusted to calculate profit in multiples of 10

        dummy_row = {
            'InvoiceDate': date.strftime('%d-%m-%Y'),
            'Retailer': retailer,
            'Region': region,
            'State': state,
            'City': city,
            'Product': product,
            'Price per Unit': price_per_unit,
            'UnitsSold': units_sold,
            'TotalSales': total_sales,
            'OperatingProfit': operating_profit,
            'OperatingMargin': operating_margin,
            'SalesMethod': sales_method
        }
        dummy_data_rows.append(dummy_row)

# Create DataFrame
sales_data = pd.DataFrame(dummy_data_rows, columns=columns)

# Save to CSV
sales_data.to_csv('sales_data.csv', index=False)
```
