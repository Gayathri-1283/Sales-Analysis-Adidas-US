```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year
```

```python
total_sales_by_state = df.groupby(['Year','State','state_code'])['TotalSales'].sum().reset_index()
```
